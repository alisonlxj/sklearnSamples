```python

# --------------------------
# here is my code
# --------------------------

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score


# 载入数据
datasets_x, datasets_y = datasets.load_diabetes(return_X_y=True)
# 只取 一个特征 (二维矩阵第二列数据)
datasets_x = datasets_x[:, np.newaxis, 2]

# 筛分 训练接和测试集
datasets_x_train = datasets_x[:-20]             # 0至倒数第20个
datasets_x_test = datasets_x[-20:]              # 最后20个作为测试集

# 对应 筛分target结果 到 训练集测试集
datasets_y_train = datasets_y[:-20]
datasets_y_test = datasets_y[-20:]


# 创建 线性回归模型对象
regr = linear_model.LinearRegression()
# 训练集训练模型
regr.fit(datasets_x_train, datasets_y_train)

# 测试集预测模型
datasets_y_predict = regr.predict(datasets_x_test)


print("The Coefficient is : ", regr.coef_)
print("The Mean Squared Error is %.2f" %mean_squared_error(datasets_y_test, datasets_y_predict))

print("The Coefficient of Determination is : %.2f" %r2_score(datasets_y_test, datasets_y_predict))


```

```python

# 画图显示

# 训练集展示
plt.scatter(datasets_x_train, datasets_y_train, color='black')

# 测试集展示
plt.scatter(datasets_x_test, datasets_y_test, color='red')
# 预测模型 展示测试结果
plt.plot(datasets_x_test, datasets_y_predict, color='blue', linewidth=3)

# plt.xticks(())
# plt.yticks(())

plt.show()
```
